```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor

df = pd.read_csv("train.csv")

df.head()
# sukses mengambil data
# pisahkan x dan y
X = df.drop(['SalePrice'], axis=1)
y = df['SalePrice']

print(X.head())
print(y.head())
```

```python
# melakukan pre processing data
# melakukan encoding dengan ordinal encoding
object_cols = X.select_dtypes(['object'])
print(object_cols.columns)
# mendapatkan apa saja objectnya
my_encoder = OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1)

print(object_cols)
```

```python
label_x = X.copy()

# pastikan menggunakan list agar tidak error
list_of_cols = list(object_cols)

# mulai encodin
label_x[list_of_cols] = my_encoder.fit_transform(X[list_of_cols])

# cetak hasil encoding
print("Hasil Dari encodingnya adalah")
print(label_x.head(10))
```

```python
# karena sudah semuanya berbentuk data
# saatnya imputasi data yang kosong
# ambil column yang ada data yang hilang
cols_with_missing = [col for col in label_x.columns if label_x[col].isnull().any()]

my_imputer = SimpleImputer(strategy='mean')
X_train_plus = label_x.copy()

print("Kolom yang terdapat hilang")
for i in cols_with_missing:
    print(i)

print("summary data yang kosong berdasarkan kuantitas")
missing_count = label_x[cols_with_missing].isnull().sum()
print(missing_count[missing_count > 0].sort_values(ascending=False))
```

```python
# setelah menemukan, maka mulai untuk input data
for col in cols_with_missing:
    X_train_plus[col + '_was_missing'] = X_train_plus[col].isnull()

# imputation
imputed_X_train = pd.DataFrame(my_imputer.fit_transform(X_train_plus))

print(imputed_X_train.head())

# samakan kembali columnsnya
imputed_X_train.columns = X_train_plus.columns
print("hasil setelah di imputasi adalah")
print(imputed_X_train.head())
```

```python
# karena sudah mendapatkan semua datanya maka lakukan pengetesan dengan split data
train_X, test_X, train_y, test_y = train_test_split(imputed_X_train, y, test_size=0.2, random_state=1)

# mencoba print dengan itu
print(train_X.head())
```

```python
# Model 1: Baseline (default-ish, buat pembanding)
model_1 = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

# Model 2: Shallow trees, banyak pohon (mengurangi overfitting)
model_2 = RandomForestRegressor(
    n_estimators=300,
    max_depth=5,
    random_state=42
)

# Model 3: Deep trees, kontrol lewat max_leaf_nodes
model_3 = RandomForestRegressor(
    n_estimators=200,
    max_leaf_nodes=50,
    random_state=42
)

# Model 4: Kontrol lewat min_samples_split & min_samples_leaf (anti overfit)
model_4 = RandomForestRegressor(
    n_estimators=200,
    min_samples_split=10,
    min_samples_leaf=4,
    random_state=42
)

# Model 5: max_features dikurangi (lebih random per split, cocok kalau fitur banyak)
model_5 = RandomForestRegressor(
    n_estimators=300,
    max_features="sqrt",
    max_depth=15,
    random_state=42
)

# Model 6: max_features "log2", pohon lebih banyak lagi
model_6 = RandomForestRegressor(
    n_estimators=500,
    max_features="log2",
    min_samples_leaf=2,
    random_state=42
)

# Model 7: Full depth (unconstrained) tapi banyak sample per leaf buat stabilin
model_7 = RandomForestRegressor(
    n_estimators=400,
    max_depth=None,
    min_samples_leaf=3,
    bootstrap=True,
    max_samples=0.8,   # tiap tree cuma pake 80% data (subsampling)
    random_state=42
)

# Model 8: Kombinasi "agresif" - banyak tree, depth sedang, feature terbatas
model_8 = RandomForestRegressor(
    n_estimators=600,
    max_depth=20,
    min_samples_split=5,
    min_samples_leaf=2,
    max_features=0.5,
    max_samples=0.9,
    random_state=42
)

from sklearn.model_selection import cross_val_score

models = {
    "model_1": model_1,
    "model_2": model_2,
    "model_3": model_3,
    "model_4": model_4,
    "model_5": model_5,
    "model_6": model_6,
    "model_7": model_7,
    "model_8": model_8,
}

results = {}

for name, model in models.items():
    # scoring pake neg RMSE (default RMSE Kaggle house pricing biasanya di log target)
    scores = cross_val_score(
        model, train_X, train_y,
        cv=5,
        scoring="neg_root_mean_squared_error",
        n_jobs=-1
    )
    rmse = -scores.mean()
    results[name] = rmse
    print(f"{name}: RMSE = {rmse:.4f} (+/- {scores.std():.4f})")

# Urutin dari yang paling bagus
best_model = min(results, key=results.get)
print(f"\nModel terbaik sementara: {best_model} dengan RMSE {results[best_model]:.4f}")
```

```python
model_8.fit(train_X, train_y)
```

```python
# memakai model_8 untuk data test
dft = pd.read_csv("test.csv")

# memulai melakukan imputasi dan sebagainya
dft_test_id = dft['Id']

# lakukan encoding kolom kategorikal
X_test_encoded = dft.copy()
good_object_cols = list(object_cols)
X_test_encoded[good_object_cols] = my_encoder.transform(dft[good_object_cols])

# lakukan imputasi missing value
# berikan beberapa parameter dulu untuk was missing
X_plus_test = X_test_encoded.copy()
for col in cols_with_missing:
    X_plus_test[col + '_was_missing'] = X_plus_test[col].isnull()
X_test_imputed = pd.DataFrame(my_imputer.transform(X_plus_test))
# samakan kolomnya kembali
X_test_imputed.columns = X_plus_test.columns

X_test_imputed.head()
```

```python
# lakukan pemodelan prediksi
prediction = model_8.predict(X_test_imputed)

print(prediction)

submission = pd.DataFrame({
    'Id': dft_test_id,
    'SalePrice': prediction
})

# Export ke file CSV tanpa menyertakan index pandas
submission.to_csv('submission.csv', index=False)
print("File submission.csv berhasil dibuat! Siap di-upload ke Kaggle.")
```
